### Thresholding Example

```python
import cv2 #image processing lib
import numpy as np #fast math
from matplotlib import pyplot as plt #plotting
```

```python
plt.rcParams['figure.figsize'] = [20, 10] #figure size
```

```python
!wget -O image_gradient.png https://drive.google.com/u/0/uc?id=1aDcdpyUcrDd174BUgeMdfLa4BQDwM0Bd&export=download
#directory = '/Users/usernanme/Documents/image_gradient.png'
img = cv2.imread('image_gradient.png') #load png data in
```

```python
plt.imshow(img) #show the image
```

```python
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY) #Binary
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV) #Binary inverted
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC) #Truncate
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO) #Threshold to zero
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV) #Threshold to zero inverted
```

```python
titles = ['Original Image','BINARY','BINARY_INVERTED','TRUNCATE','TO_ZERO','TOZERO_INVERTED'] #Plot titles
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5] #raw and processed image data
```

```python
for i in range(len(images)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    
    plt.title(titles[i])
    plt.axis('off')
    
plt.tight_layout()
plt.show()
```
